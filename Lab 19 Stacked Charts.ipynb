### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
##Write your code here
# Group by Age and calculate median values
df_stack1 = df[['Age', 'JobSatPoints_6', 'JobSatPoints_7']].dropna()
df_grouped1 = df_stack1.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Plot
df_grouped1.plot(kind='bar', stacked=True, figsize=(12, 6), title="Median Job Satisfaction Scores by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Median Job Satisfaction Score")
plt.tight_layout()
plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
##Write your code here
# Group by Employment and calculate mean scores
df_stack2 = df[['Employment', 'JobSatPoints_6', 'JobSatPoints_7']].dropna()
df_emp_clean['Employment'] = df_stack2['Employment'].str.split(';')
df_emp_clean = df_emp_clean.explode('Employment')
df_emp_clean['Employment'] = df_emp_clean['Employment'].str.strip()  # remove whitespace

# Group and plot after cleaning
df_grouped_clean = df_emp_clean.groupby(['JobSat', 'Employment']).size().unstack(fill_value=0)

# Plot
df_grouped_clean.plot(kind='bar', stacked=True, figsize=(12, 6), title="Job Satisfaction by Employment Status")
plt.xlabel("Employment Status")
plt.ylabel("Average Job Satisfaction Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
##Write your code here
# Split databases and explode
df_db_age = df[['Age', 'DatabaseWantToWorkWith']].dropna()
df_db_age = df_db_age.assign(Database=df_db_age['DatabaseWantToWorkWith'].str.split(';')).explode('Database')

# Count frequency per age/database
df_grouped5 = df_db_age.groupby(['Age', 'Database']).size().unstack(fill_value=0)

# Plot
df_grouped5.plot(kind='bar', stacked=True, figsize=(16, 12), title="Preferred Databases by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count of Preferences")
plt.tight_layout()
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
##Write your code here
# Clean and explode Employment field
df_emp_clean = df[['Employment', 'JobSat']].dropna().copy()
df_emp_clean['Employment'] = df_emp_clean['Employment'].str.split(';')
df_emp_clean = df_emp_clean.explode('Employment')
df_emp_clean['Employment'] = df_emp_clean['Employment'].str.strip()  # remove whitespace

# Group and plot after cleaning
df_grouped_clean = df_emp_clean.groupby(['JobSat', 'Employment']).size().unstack(fill_value=0)

# Plot
df_grouped_clean.plot(kind='bar', stacked=True, figsize=(14, 7), title="Cleaned Employment Type by Job Satisfaction")
plt.xlabel("Job Satisfaction")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
##Write your code here
# Split admired languages and explode
df_lang_age = df[['Age', 'LanguageAdmired']].dropna()
df_lang_age = df_lang_age.assign(Language=df_lang_age['LanguageAdmired'].str.split(';')).explode('Language')

# Count per age/language
df_grouped7 = df_lang_age.groupby(['Age', 'Language']).size().unstack(fill_value=0)

# Plot
df_grouped7.plot(kind='bar', stacked=True, figsize=(16, 20), title="Admired Programming Languages by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Language Admiration Count")
plt.tight_layout()
plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
##Write your code here
# Split admired platforms and explode
df_platform_emp = df[['Employment', 'PlatformAdmired']].dropna()
df_platform_emp = df_platform_emp.assign(Platform=df_platform_emp['PlatformAdmired'].str.split(';')).explode('Platform')
df_platform_emp = df_platform_emp.assign(Employment = df_platform_emp['Employment'].str.split(';')).explode('Employment')

# Count per employment/platform
df_grouped8 = df_platform_emp.groupby(['Employment', 'Platform']).size().unstack(fill_value=0)

# Plot
df_grouped8.plot(kind='bar', stacked=True, figsize=(16, 20), title="Admired Platforms by Employment Type")
plt.xlabel("Employment Type")
plt.ylabel("Admiration Count")
plt.tight_layout()
plt.show()

```
